```python
import pandas as pd
```

```python
df = pd.read_csv('covid_data_set.csv', error_bad_lines=False)
```

```python
n_df = df[["symptoms"]]
```

```python
sym=n_df.dropna(axis='rows')
```

```python
def check_fever(x):
    if "fever" in x:
        return 1
    else:
        return 0  
def check_fatigue(x):
    if "fatigue" in x:
        return 1
    else:
        return 0 
def check_pneumonia(x):
    if "pneumonia" in x:
        return 1
    else:
        return 0 
def check_bone_pain(x):
    if "bone pain" in x:
        return 1
    else:
        return 0 
def check_cough(x):
    if "cough" in x:
        return 1
    else:
        return 0 
def check_breath(x):
    if "breath" in x:
        return 1
    else:
        return 0 
def check_diarrheoa(x):
    if "diarrheoa" in x:
        return 1
    else:
        return 0
def check_headache(x):
    if "headache" in x:
        return 1
    else:
        return 0 
```

```python
sym['fever']=sym['symptoms'].apply(check_fever)
sym['fatigue']=sym['symptoms'].apply(check_fatigue)
sym['pneumonia']=sym['symptoms'].apply(check_pneumonia)
sym['headache']=sym['symptoms'].apply(check_headache)
sym['diarrheoa']=sym['symptoms'].apply(check_diarrheoa)
sym['cough']=sym['symptoms'].apply(check_cough)
sym['bone_pain']=sym['symptoms'].apply(check_bone_pain)
```

```python
sym.to_csv("corona_symptoms.csv",index=False)
```
